```python
#collaborative filtering is the field in which we find similarities between different user by their buying preference,ratings,
#likes and dislike soo that in future we if one buy/watch something and give good ratings then we can recommend same to other
import pandas as pd
dataset=pd.read_csv('ratings.csv')
```

```python
#if in case our dataset is large and we want to see few dataset the we use head function
dataset.head(5)
```

```python
#to remove any column from dataset we use drop and in case if we want to remove from original file inplace
#if we want to remove column from the dataframe
dataset.drop('timestamp',axis=1,inplace=True)
```

```python
dataset.head(5)
```

```python
#to find the number of unique user we use len and unique
len(dataset.userId.unique())
```

```python
len(dataset.movieId.unique())
```

```python
#to convert the dataframe according to our need we use pivot index for row and columns for column and values for value int he table
user_movie_df=dataset.pivot(index='userId',columns='movieId',values='rating')
```

```python
user_movie_df.iloc[0:5,0:10]
```

```python
#if our dataset contain Nan value and we want to replace it with 0 we use fillna
user_movie_df.fillna(0,inplace=True)
```

```python
user_movie_df.iloc[0:5,0:10]
```

```python
#to find the distance from user with every user we use pairwise distance first pass values and then distance in this case we cosine distance
from sklearn.metrics import pairwise_distances
user_sim=1-pairwise_distances(user_movie_df.values,metric='cosine')
```

```python
user_sim
```

```python
#our dataset is in the array from and we want to convert it into dataframe so we use DataFrame and the value pass will be
#array,row,columns
user_sim_df=pd.DataFrame(user_sim,index=dataset.userId.unique(),columns=dataset.userId.unique())
```

```python
user_sim_df
```

```python
#to convert the diagonal value zero we use fill_diagonal
import numpy as np
np.fill_diagonal(user_sim,0)
```

```python
#to find the max distance in a row we are just finding the similar user
user_sim_df.idxmax(axis=1)
```

```python
#importing the dataset
import pandas as pd
dataset=pd.read_csv('ratings.csv')
```

```python
dataset.head()
```

```python
#droping the column from the dataframe through drop by giving columns name and inplace  
dataset.drop(columns='timestamp',inplace=True)
```

```python
#changing the dataframe by making row as movie id and column as userid and values are rating in them
movie_user_df=dataset.pivot(index='movieId',columns='userId',values='rating')
```

```python
movie_user_df.head()
```

```python
#if the dataframe contains Nan value and if we want to remove them and place some vlue there then that can be done by fillna
movie_user_df.fillna(0,inplace=True)
movie_user_df.head()
```

```python
#we are here finding the cosine similarity
from sklearn.metrics import pairwise_distances
movie_sim=1-pairwise_distances(movie_user_df.values,metric='cosine')
```

```python
#if we want to change the diagonal value then we can do this by nupy library filldiagonal by providing numpy array and value to be fill
import numpy as np
np.fill_diagonal(movie_sim,0)
```

```python
#creating dataframe of the movie with respect to their cosine similarity
movie_sim_df=pd.DataFrame(movie_sim,index=dataset.movieId.unique(),columns=dataset.movieId.unique())
movie_sim_df.head()
```

```python
#by this we are just finding the two most similar value
movie_sim_df.idxmax(axis=1)
```
